* NUMBER 6: Whate are the peak hour of sales?

```python
sales_sum = df.groupby('time')['price'].sum()  # total sales par hour
sales_sum 
```

## NUMBER 7: Find the total sale made on each day of the week. which day of the week is when sales are made the most

* TO GET THIS DONE, I NEED TO EXTRACT DAYS OF THE WEEK FROM DATE COLUMN.. TO GET ANOTHER COLUMN FOR DAY OF THE WEEK.

```python
df
```

```python
df['day_of_week'] = df['date'].dt.day_name()
df
```

```python
sales_par_day = df.groupby('day_of_week')['Revenue'].sum()
print(sales_par_day)
```

```python
Highest_sales_day = sales_par_day.idxmax()
print('Day With The Highest Sales:', Highest_sales_day)
```

```python
top3_highest_sales_days = sales_par_day.sort_values(ascending= False).head(3)
print('\nTop 3 highest sales days :')
print(top3_highest_sales_days)
```

* NUMBER 9:Find the sales made in each moth. Any trend noticeable?

```python
df['month'] = df['date'].dt.month_name()
df['month_number'] = df['date'].dt.month # extract months and month number from the date column
df
```

```python
monthly_sales = (
    df.groupby(['month_number', 'month'])['Revenue']
    .sum()
    .reset_index()
    .sort_values('month_number')
)

print(monthly_sales[['month', 'Revenue']])
```

```python
plt.figure()
plt.plot(monthly_sales['month'], monthly_sales['Revenue'])
plt.xticks(rotation=45)
plt.title("Total Sales Per Month")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```
